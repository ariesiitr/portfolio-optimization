READING THE CSV FILES INTO A DATAFRAME

```python
voo=pd.read_csv(csv_file_path1)
blv=pd.read_csv(csv_file_path2)
voo.head()
```

```python
blv.head()
```

```python
# Assuming "Adj Close" column exists in both blv and voo DataFrames
blv["return"] = (blv["Adj Close"] - blv["Adj Close"].shift(1)) / blv["Adj Close"]
voo["return"] = (voo["Adj Close"] - voo["Adj Close"].shift(1)) / voo["Adj Close"]
```

CALCULATING THE DAILY RETURNS

```python
voo_mean=voo["return"].mean()
blv_mean=blv["return"].mean()
voo_var=voo["return"].var()
blv_var=blv["return"].var()
voo_stdev=math.sqrt(voo_var)
blv_stdev=math.sqrt(blv_var)
print(voo_mean)
```

```python
voo_sharpe_ratio=(voo_mean-0.19)/voo_stdev
blv_sharpe_ratio=(blv_mean-0.19)/blv_stdev
covariance = blv['return'].cov(voo['return'])
correlation=blv['return'].corr(voo['return'])
print(covariance,correlation)
```

CALCULATING THE VARIABLES FOR FURTHER CALCULATIONS

```python
MEAN=[]
VAR=[]
STDEV=[]
SHARPE=[]
voo_percentage_invested=[]
blv_percentage_invested=[]
for i in range(101):
    MEAN.append((i*voo_mean)+(100-i)*blv_mean)
    VAR.append(((i*i)*voo_var)+(((100-i)*(100-i))*blv_var)+(2*i*(100-i)*covariance))
    STDEV.append(math.sqrt(VAR[i]))
    SHARPE.append((MEAN[i]-0.19)/STDEV[i])
    voo_percentage_invested.append(i)
    blv_percentage_invested.append(100-i)

print(MEAN)
```

```python
df = pd.DataFrame({'MEAN': MEAN, 'STDEV': STDEV , 'SHARPE':SHARPE,'voo_percentage_invested':voo_percentage_invested,'blv_percentage_invested':blv_percentage_invested})


fig = px.scatter(
    df, 
    x='STDEV', 
    y='MEAN', 
    hover_data={'SHARPE': True,'MEAN':False,'STDEV':False,'voo_percentage_invested':True,'blv_percentage_invested':True},
    labels={'MEAN': 'REWARD ', 'STDEV': 'RISK '},
    title='THE MARKOWITZ MODEL'
)


fig.show()
```
